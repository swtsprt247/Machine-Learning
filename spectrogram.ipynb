## Mel Spectrogram

Mel spectrograms extract the frequency information over the course of the song, and display how loud or prominent each frequency is. With this pop song, you can see a large range of frequencies, as pop songs are typically much higher pitched than other genres. https://machineramblings.wordpress.com/2017/05/09/shazam-but-magic/

<img src="spectrogram.png">

```python
import os
import matplotlib
import pylab
import librosa
import librosa.display
import numpy as np
```

```python
sig, fs = librosa.load('summer_breeze.wav')
```

```python
pylab.axes([0., 0., 1., 1.], frameon=False, xticks=[], yticks=[]) # Remove the white edge
S = librosa.feature.melspectrogram(y=sig, sr=fs)
librosa.display.specshow(librosa.power_to_db(S, ref=np.max))
```
