```python
def RegressionPlot(X, Y, model):
    plt.figure(figsize=(8,8))
    plt.scatter(X, Y, c='steelblue', edgecolor='white', s=70)
    plt.plot(X, model.predict(X), color='black', lw=2)
    return 
```

```python
from sklearn.ensemble import GradientBoostingRegressor

X = music_data[['tempo']].values
y = music_data['song.hotttnesss'].values


from sklearn.linear_model import LinearRegression

model = GradientBoostingRegressor(random_state=12, n_estimators=1000, max_depth=4)
model.fit(X, y)

sort_idx = X.flatten().argsort()
RegressionPlot(X[sort_idx], y[sort_idx], model)
plt.show()
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=12)
model = GradientBoostingRegressor(n_estimators=3000, max_depth=6, learning_rate=0.01,
                                loss='huber', random_state=0, verbose=1)
model.fit(X_train,y_train)
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
```

```python
parameter_grid = {
                 'max_depth' : [4, 6, 8, 10],
                 'n_estimators': [50, 10, 20, 100, 200],
                 'max_features': ['sqrt', 'auto', 'log2'],
                 'min_samples_split': [2, 3,4,5,6,7],
                 'min_samples_leaf': [2,3,4,5],
                 }
boost = GradientBoostingRegressor()
cross_validation = StratifiedKFold(n_splits=5)

grid_search = GridSearchCV(boost,
                              
                               param_grid=parameter_grid,
                               cv=2,
                               verbose=1,
                               n_jobs=-1
                              )

grid_search.fit(X_train, y_train)
model = grid_search
parameters = grid_search.best_params_

print('Best score: {}'.format(grid_search.best_score_))
print('Best parameters: {}'.format(grid_search.best_params_))
```
